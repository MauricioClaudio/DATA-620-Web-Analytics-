```python
### Import libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
### Magical command for matplotlib to print inline
%matplotlib inline

### Instantiate graph
grp=nx.Graph()

#### Define nodes
# Andre
grp.add_edge('Andre','Beverly')
grp.add_edge('Andre','carol')
grp.add_edge('Andre','diane')

# Beverly
grp.add_edge('Beverly','carol')
grp.add_edge('Beverly','diane')
grp.add_edge('Beverly','ed')

# Carol
grp.add_edge('carol','Andre')
grp.add_edge('carol','diane')
grp.add_edge('carol','fernando')

# Fernando
grp.add_edge('fernando','garth')
grp.add_edge('fernando','diane')
grp.add_edge('fernando','carol')

# Garth
grp.add_edge('garth','ed')
grp.add_edge('garth','diane')
grp.add_edge('garth','fernando')
grp.add_edge('garth','heather')

# Ed
grp.add_edge('ed','Beverly')
grp.add_edge('ed','diane')
grp.add_edge('ed','garth')

# Heather
grp.add_edge('heather','fernando')
grp.add_edge('heather','garth')
grp.add_edge('heather','ike')

# Ike
grp.add_edge('ike','heather')
grp.add_edge('ike','Jane')

# Jane
grp.add_edge('Jane','ike')
```

```python
### Define node positions 
pos={
    'Andre':(0,1),
    'Beverly':(0,-1),
    'carol':(1,2),
    'diane':(1,0),
    'ed':(1,-2),
    'fernando':(2,1),
    'garth':(2,-1),
    'heather':(3,0),
    'ike':(4,0),
    'Jane':(5,0)
}

### Node params
nx.draw_networkx_nodes(grp, pos,
                       node_size=2100,
                       node_color='dodgerblue',
                       alpha=0.7)

### Edge params
nx.draw_networkx_edges(grp, pos,
                       alpha=0.5,
                       width=5)

### label params
nx.draw_networkx_labels(grp, pos,
                        font_size=12)

### Plot the thing
plt.axis('off')
plt.show() # display
```
